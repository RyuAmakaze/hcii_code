```python
from numpy.random import *
import numpy as np 
import scipy.stats as stats
import pandas as pd
import csv
import matplotlib.pyplot as plt
```

```python
name = "test"

f = open("data/"+name+"/alpha.csv", "r")
csv_data = csv.reader(f)
data = [ e for e in csv_data]
f.close()

ab = np.zeros(0)
for i in range(30):
    ab = np.append(ab,float(data[0][i]))
    
print(ab)
print(stats.shapiro(ab))
```

```python
plt.hist(ab, bins=20)
stats.shapiro(ab)
```

```python
stats.probplot(ab, dist="norm",plot=plt)
```
